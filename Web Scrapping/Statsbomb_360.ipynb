```python
Kobbie=99174

df=df[(df["player_id"]==Kobbie) & (df["type"]=="Pass")].reset_index(drop=True)
```

```python
df.location
```

```python
df[["x_start","y_start"]]=pd.DataFrame(df.location.tolist(),index=df.index)
df[["x_end","y_end"]]=pd.DataFrame(df.pass_end_location.tolist(),index=df.index)
```

```python
fig, ax= plt.subplots(figsize=(10.5,6))
fig.set_facecolor("#22312b")
ax.patch.set_facecolor("#22312b")

pitch = Pitch(pitch_type='statsbomb', pitch_color='#22312b', line_color='#c7d5cc')
pitch.draw(ax=ax, constrained_layout=True, tight_layout=False)
plt.gca().invert_yaxis()

for x in range(len(df["location"])):
    pitch.scatter(x=df["x_start"][x],y=df["y_start"][x],ax=ax,color="white")
    pitch.lines(xstart=df["x_start"][x], xend=df["x_end"][x],ystart=df["y_start"][x],yend=df["y_end"][x],ax=ax,color="white",comet=True,alpha=0.5)

plt.title("Kobbie Mainoo Passmap vs Spain Euro Finals 2025",color="white")
```

```python
fig, ax= plt.subplots(figsize=(10.5,6))
fig.set_facecolor("#22312b")
ax.patch.set_facecolor("#22312b")

pitch = Pitch(pitch_type='statsbomb', pitch_color='#22312b', line_color='#c7d5cc')
pitch.draw(ax=ax, constrained_layout=True, tight_layout=False)
plt.gca().invert_yaxis()

df=df[1:2]

pitch.scatter(x=df["x_start"],y=df["y_start"],ax=ax,color="white")
pitch.lines(xstart=df["x_start"], xend=df["x_end"],ystart=df["y_start"],yend=df["y_end"],ax=ax,color="white",comet=True,alpha=0.5)

for x in df.iloc[0]["freeze_frame"]:
    if x["teammate"]:
        color="white"
    else:
        color="red"
    pitch.scatter(x=x["location"][0],y=x["location"][1],ax=ax,c=color)
```
