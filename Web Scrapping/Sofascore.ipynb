```python
import requests

headers = {
    'accept': '*/*',
    'accept-language': 'en-US,en;q=0.9',
    'referer': 'https://www.sofascore.com/football/match/moldova-norway/AOb5Hv',
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/139.0.0.0 Safari/537.36',
}

url = "https://www.sofascore.com/api/v1/event/13233577/shotmap"

response = requests.get(url, headers=headers)

print(response.status_code) 
```

```python
shots=response.json()
```

```python
df = pd.json_normalize(shots["shotmap"])
```

```python
df.head()
```

```python
df.columns
```

```python
df
```

```python
df=df[["shotType","situation","bodyPart","time","player.shortName","player.jerseyNumber","playerCoordinates.x","playerCoordinates.y","isHome"]]
df["playerCoordinates.x"]=df["playerCoordinates.x"]*1.2
df["playerCoordinates.y"]=df["playerCoordinates.y"]*0.8
```

```python
df
```

```python
fig, ax=plt.subplots(figsize=(10.5,7))
fig.set_facecolor("#22312b")
ax.patch.set_facecolor("#22312b")

pitch = VerticalPitch(pitch_type='statsbomb', pitch_color='#22312b', line_color='#c7d5cc',half=True)
pitch.draw(ax=ax, constrained_layout=True, tight_layout=False )
plt.gca().invert_yaxis()

ax.set_xlim(0,80)
ax.set_ylim(0,45)

for x in range(len(df["playerCoordinates.x"])):
    if (df["isHome"][x]==True):
        if df["shotType"][x]=="save":
            plt.scatter(y=df["playerCoordinates.x"][x],x=df["playerCoordinates.y"][x],color="#BA0C2F",s=100,marker="^")
        if df["shotType"][x]=="miss":
            plt.scatter(y=df["playerCoordinates.x"][x],x=df["playerCoordinates.y"][x],color="#BA0C2F",s=100,marker="X")
        if df["shotType"][x]=="block":
            plt.scatter(y=df["playerCoordinates.x"][x],x=df["playerCoordinates.y"][x],color="#BA0C2F",s=100,marker="s")
        if df["shotType"][x]=="goal":
            plt.scatter(y=df["playerCoordinates.x"][x],x=df["playerCoordinates.y"][x],color="#BA0C2F",s=100,marker="*")
    if (df["isHome"][x]==False):
        if df["shotType"][x]=="save":
            plt.scatter(y=df["playerCoordinates.x"][x],x=df["playerCoordinates.y"][x],color="#FFD200",s=100,marker="^")
        if df["shotType"][x]=="miss":
            plt.scatter(y=df["playerCoordinates.x"][x],x=df["playerCoordinates.y"][x],color="#FFD200",s=100,marker="X")
        if df["shotType"][x]=="block":
            plt.scatter(y=df["playerCoordinates.x"][x],x=df["playerCoordinates.y"][x],color="#FFD200",s=100,marker="s")
        if df["shotType"][x]=="goal":
            plt.scatter(y=df["playerCoordinates.x"][x],x=df["playerCoordinates.y"][x],color="#FFD200",s=100,marker="*")

plt.title("Norway vs Moldova Shotmap",color="white",size="20")
ax.text(4,40,"Goal",c="white")
ax.scatter(8,40.5,marker="*",s=100)
ax.text(4,38,"Off Target",color="white")
ax.scatter(12,38.5,marker="X",s=100)
ax.text(4,36,"Saved",color="white")
ax.scatter(9.5,36.5,marker="^",s=100)
ax.text(4,34,"Blocked",color="white")
ax.scatter(10.5,34.5,marker="s",s=100)

ax.text(70,40,"Norway",c="#BA0C2F")
ax.text(70,38,"Moldova",c="#FFD200")

plt.savefig("C:/Users/K Raghunandan/Analyst/Results/Norway vs Moldova 2025")
```
