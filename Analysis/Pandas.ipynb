```python
import pandas as pd
```

```python
data={
    "Opponent":["Atletico Jave","Newton FC","Bunton Town","Fentborough Dynamo"],
    "Location":["Home","Away","Away","Home"],
    "Goals For(GF)":[2,4,3,0],
    "Goals Against(GA)":[4,0,2,2],}
Matches=pd.DataFrame(data)
Matches["Goal Difference(GD)"]= Matches["Goals For(GF)"]-Matches["Goals Against(GA)"]
Matches.loc[Matches["Goal Difference(GD)"]>0, "Result"] = "W"
Matches.loc[Matches["Goal Difference(GD)"]==0, "Result"] = "D"
Matches.loc[Matches["Goal Difference(GD)"]<0, "Result"] = "L"
Matches
```

```python
Matches.groupby("Location").describe()
```

```python
import numpy as np
import pandas as pd
```

```python
match1 = pd.DataFrame({
    "Opponent":["Selche FC"],
    "Goals For(GF)":[1],
    "Goals Against(GA)":[1],
    "Attendance":[53225]})
match2 = pd.DataFrame({
    "Opponent":["Sudaton FC"],
    "Goals For(GF)":[3],
    "Goals Against(GA)":[0],
    "Attendance":[53256]})
match3 = pd.DataFrame({
    "Opponent":["Ouestjambon United"],
    "Goals For(GF)":[4],
    "Goals Against(GA)":[1],
    "Attendance":[53225]})
match3
```

```python
AllMatches=pd.concat([match1,match2,match3])
AllMatches
```

```python
match1scorers=pd.DataFrame({
    "First":["Sally"],
    "Last":["Billy"],
    "Opponent":["Selche FC"]})
match2scorers=pd.DataFrame({
    "First":["Sally"],
    "Last":["Pip"],
    "Opponent":["Sudaton FC"]})
match3scorers=pd.DataFrame({
    "First":["Sally"],
    "Last":["Silly"],
    "Opponent":["Ouestjambon United"]})
AllScorers=pd.concat([match1scorers,match2scorers,match3scorers])
AllScorers
```

```python
pd.merge(AllMatches,AllScorers,how="inner",on="Opponent")
```

```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("C:/Users/K Raghunandan/OneDrive/Desktop/Raghu/Kickoff/Results.csv")
df.head()
```

```python
print(df["HomeTeam"].unique())
print(str(len(df["HomeTeam"].unique()))+" Unique Teams!")
```

```python
len(df)
```

```python
df.columns
```

```python
del df["Referee"]
```

```python
df["Result"]=df["FTHG"]-df["FTAG"]
```

```python
def findWinner(value):
    if (value>0):
        return ("Home Win")
    elif (value<0):
        return ("Away Win")
    else:
        return ("Draw")
df["ResultText"]=df["Result"].apply(findWinner)
df.head()
```

```python
print(df["FTHG"].mean())
print(df["FTAG"].mean())
```

```python
df["TotalGoals"]=df["FTHG"]+df["FTAG"]
```

```python
df.groupby("HomeTeam").mean()["TotalGoals"].sort_values(ascending=False)
```
