```python
df2=pd.read_csv("C:/Users/K Raghunandan/OneDrive/Desktop/Raghu/Kickoff/CSV Files/beeswarm2.csv")
df2.head()
```

```python
df2["Player"]=df2["Player"].str.split("\\",expand=True)[0]
```

```python
df2=df2[df2["90s"]>=6.5]
df2=df2[df2["Pos"]!="GK"].reset_index()
df2
```

```python
#Progression
df2["Prog"]=df2["Prog"]/df2["90s"]
#Passes into Final Third
df2["1/3"]=df2["1/3"]/df2["90s"]
#Expected Assists
df2["xA"]=df2["xA"]/df2["90s"]
#Key PAsses
df2["KP"]=df2["KP"]/df2["90s"]
#Passes into Penalty Area
df2["PPA"]=df2["PPA"]/df2["90s"]
```

```python
df2
```

```python
metrics=["Prog","1/3","xA","Cmp%","KP","PPA"]
```

```python
fig,axes=plt.subplots(3,2,figsize=(14,10))
fig.set_facecolor(bg)
#ax.patch.set_facecolor(bg)

mpl.rcParams["xtick.color"]=tc
mpl.rcParams["ytick.color"]=tc

counter=0
counter2=0
met_counter=0

#create initial for loop
for i,ax in zip(df2["Player"],axes.flatten()):
    ax.set_facecolor(bg)
    spines=["top","bottom","left","right"]
    for x in spines:
        if x in spines:
            ax.spines[x].set_visible(False)
    sns.swarmplot(x=metrics[met_counter],data=df2,ax=axes[counter,counter2],zorder=1)
    ax.set_xlabel(f"{metrics[met_counter]}",color="white")

    for x in range(len(df2["Player"])):
        if df2["Player"][x]=="Bruno Fernandes":
            ax.scatter(x=df2[metrics[met_counter]][x],y=0,s=200,color="red",zorder=2)
        if df2["Player"][x]=="Kevin De Bruyne":
            ax.scatter(x=df2[metrics[met_counter]][x],y=0,s=200,color="#6CADDF",zorder=2)

    met_counter+=1
    if counter2==0:
        counter2=1
        continue
    if counter2==1:
        counter2=0
        counter+=1
```
