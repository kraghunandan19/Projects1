```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from mplsoccer.pitch import Pitch
```

```python
df=pd.read_csv("C:/Users/K Raghunandan/OneDrive/Desktop/Raghu/Kickoff/xgtutorial.csv")
df
```

```python
a_xG=[0]
h_xG=[0]
a_min=[0]
h_min=[0]

hteam=df["team"].iloc[0]
ateam=df["team"].iloc[-1]

for x in range(len(df["xG"])):
    if df["team"][x]==ateam:
        a_xG.append(df["xG"][x])
        a_min.append(df["minute"][x])

    if df["team"][x]==hteam:
        h_xG.append(df["xG"][x])
        h_min.append(df["minute"][x])
```

```python
def nums_cumulative_sum(nums_list):
    return [sum(nums_list[:i+1]) for i in range(len(nums_list))]

a_cum=nums_cumulative_sum(a_xG)
h_cum=nums_cumulative_sum(h_xG)
```

```python
a_total=round(a_cum[-1],2)
h_total=round(h_cum[-1],2)
```

```python
fig,ax=plt.subplots(figsize=(10,6))
fig.set_facecolor("#3d4849")
ax.patch.set_facecolor("#3d4849")

plt.xticks([0,15,30,45,60,75,90])
plt.xlabel("Minutes")
plt.ylabel("xG")

ax.step(x=a_min,y=a_cum,where="post")
ax.step(x=h_min,y=h_cum,where="post")

plt.title("Bournemouth vs Southampton xG Flow Chart", color="white")
```
