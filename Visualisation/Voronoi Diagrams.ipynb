```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from mplsoccer.pitch import Pitch
from mplsoccer.pitch import VerticalPitch
```

```python
barca_x=[93,96,99,102,104,95,86,79,89]
barca_y=[9,22,33,44,56,63,52,33,22]
barca=list(np.zeros((9,),dtype=int))

opp_x=[99,100,102,104,104,102,97,98,94,77]
opp_y=[14,22,33,43,46,55,37,27,21,45]
opp=list(np.ones((10,),dtype=int))

df=pd.DataFrame({
    "x":barca_x+opp_x,
    "y":barca_y+opp_y,
    "team":barca+opp
})

points=np.column_stack((df.x,df.y))
```

```python
fig, ax=plt.subplots(figsize=(10.5,6))
pitch = Pitch(pitch_type='statsbomb', pitch_color='black', line_color='white',half=True,axis=True,label=True,tick=True)
pitch.draw(ax=ax, constrained_layout=False, tight_layout=False)

pitch.scatter(93,9,ax=ax)
```

```python
fig, ax= plt.subplots(figsize=(10.5,6))
fig.set_facecolor("#38383b")
ax.patch.set_facecolor("#38383b")

pitch = VerticalPitch(pitch_type='statsbomb', pitch_color='#38383b', line_color='white',half=True)
pitch.draw(ax=ax, constrained_layout=False, tight_layout=True)

x=df.x
y=df.y
team1,team2=pitch.voronoi(x,y,df.team)

t1=pitch.polygon(team1,ax=ax,facecolor="red",edgecolor="white",linewidth=3,alpha=0.4)
t2=pitch.polygon(team2,ax=ax,facecolor="blue",edgecolor="white",linewidth=3,alpha=0.4)

plt.title("Barcelona vs Cadiz Voronoi diagram",color="white")
```
