```python
#Scatter Plots
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
%matplotlib inline
```

```python
table=pd.read_csv("C:/Users/K Raghunandan/OneDrive/Desktop/Raghu/Kickoff/1617table.csv")
table.head()
```

```python
fig, ax=plt.subplots()
```

```python
ax.scatter(table["GF"],table["GA"])
```

```python
ax.set_title("Goals For & Against")
ax.set_xlabel("Goals For")
ax.set_ylabel("Goals Against")
```

```python
ax.plot([table["GF"].mean(),table["GF"].mean()],[90,20],"k-",linestyle=":",lw=1)
ax.plot([90,20],[table["GA"].mean(),table["GA"].mean()],"k-",linestyle=":",lw=1)
```

```python
ax.text(18,90,"Poor Attack,Poor Defence",color="red",size="8")
ax.text(67,20,"Strong Attack, Strong Defence",color="red",size="8")
```

```python
#Pie Charts
```

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
leagueWins={
    "Team":["Manchester United","Blackburn Rovers","Arsenal","Chelsea","Manchester City","Leicester City","Liverpool FC"],
    "Championships":[13,1,3,5,10,1,2]
}
df=pd.DataFrame(leagueWins,columns=["Team","Championships"])
df
```

```python
plt.tight_layout()
```

```python
teamColours=["#f40206","#0560b5","#ce0000","#1125ff","#28cdff","#091ebc","#000000"]
```

```python
plt.pie(df["Championships"],
        labels=df["Team"],
        colors=teamColours,
        startangle=90)
```

```python
plt.title("Premier League Titles")
plt.tight_layout()
```

```python
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import squarify
import matplotlib
```

```python
data=pd.read_csv("C:/Users/K Raghunandan/OneDrive/Desktop/Raghu/Kickoff/ManCityATT.csv")
data.head()
```

```python
dataGoals=data[data["G"]>0]
```

```python
norm=matplotlib.colors.Normalize(vmin=min(dataGoals.G),vmax=max(dataGoals.G))
colors=[matplotlib.cm.Blues(norm(value)) for value in dataGoals.G]
```

```python
fig=plt.gcf()
ax=fig.add_subplot()
fig.set_size_inches(16,4.5)
```

```python
squarify.plot(label=dataGoals.Player,sizes=dataGoals.G,color=colors,alpha=.6)
```
